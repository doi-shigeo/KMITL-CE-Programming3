# Question 1: Baker's Transformation (1 dimension)

Baker's Transformation is one of the chaotic systems, that behaves like a non-linear system.
The 1 dimension version of Baker's transformation is represented as a number sequence:

$$
a_{n+1} = 
\begin{cases}
2a_{n} & (0.0 \le a_{n} \le 0.5) \\
2(1-a_{n}) & (0.5 \le a_{n} \le 1)\\
\end{cases}
$$

We assume $ a_{1} $, the initial value of number sequence, is set to 0.4 .


The code to calculate $ \{a_{n}\} $ is below for $ 1 \le n \le 60$. 
Run the program below:

````python
an = 0.4

for i in range(1, 61):
  print(i, an)
  if an <= 0.5:
    an *= 2
  else:
    an = 2 * (1 - an)


````
